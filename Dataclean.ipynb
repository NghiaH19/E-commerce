Xem dataset

```python
import pandas as pd

cot_can_lay = ['event_time', 'event_type', 'product_id', 'category_code', 'brand', 'price']

def xu_ly_va_co_dac(file_path):
    print(f"Đang xử lý file {file_path} (Quá trình này có thể mất 2-4 phút)...")
    df = pd.read_csv(file_path, usecols=cot_can_lay)
    df = df.dropna(subset=['category_code', 'brand'])

    df['date'] = df['event_time'].str[:10]

    df_grouped = df.groupby(['date', 'product_id', 'category_code', 'brand', 'price', 'event_type']).size().unstack(fill_value=0).reset_index()
    df_grouped = df_grouped.rename(columns={'view': 'luot_view', 'cart': 'luot_cart', 'purchase': 'luot_mua'})

    for col in ['luot_view', 'luot_cart', 'luot_mua']:
        if col not in df_grouped.columns:
            df_grouped[col] = 0

    print(f"-> Xử lý xong! File teo nhỏ xuống còn {len(df_grouped)} dòng.\n")
    del df
    return df_grouped

df_thang_10 = xu_ly_va_co_dac('ecommerce_data/2019-Oct.csv')
df_thang_11 = xu_ly_va_co_dac('ecommerce_data/2019-Nov.csv')

print("Đang gộp 2 tháng lại với nhau...")
df_final = pd.concat([df_thang_10, df_thang_11], ignore_index=True)

df_final = df_final.sort_values(by=['date', 'product_id']).reset_index(drop=True)

print("ĐÂY LÀ BỘ DỮ LIỆU CHUẨN ĐỂ ĐƯA VÀO MÔ HÌNH:")
display(df_final.head())
```

```python
import pandas as pd
import numpy as np

print("1. Đang gộp triệt để các dòng trùng lặp (Aggregate)...")

# Chuyển cột date về datetime chuẩn
df_final['date'] = pd.to_datetime(df_final['date'])

# Gộp các dòng trùng Product_ID và Date
# Cộng dồn lượt view/mua, lấy giá trung bình, lấy category/brand đầu tiên gặp được
df_fixed = df_final.groupby(['product_id', 'date']).agg({
    'luot_view': 'sum',
    'luot_mua': 'sum',
    'luot_cart': 'sum',
    'price': 'mean',
    'category_code': 'first',
    'brand': 'first'
}).reset_index()

print(f"   -> Sau khi gộp trùng: {len(df_fixed):,} dòng.")

# 2. Định nghĩa hàm lấp đầy ngày trống (Reindex)
def fill_missing_dates(grp):
    # Tạo dải ngày liên tục từ ngày đầu đến ngày cuối của sản phẩm đó
    all_dates = pd.date_range(start=grp['date'].min(), end=grp['date'].max(), freq='D')

    # Thiết lập index là ngày để reindex
    grp = grp.set_index('date').reindex(all_dates)

    # Những ngày mới chèn vào thì lượt view/mua/cart = 0
    grp[['luot_view', 'luot_mua', 'luot_cart']] = grp[['luot_view', 'luot_mua', 'luot_cart']].fillna(0)

    # Những thông tin cố định như Brand, Category, Price thì lấy dữ liệu gần nhất (Forward Fill)
    grp[['product_id', 'category_code', 'brand', 'price']] = grp[['product_id', 'category_code', 'brand', 'price']].ffill().bfill()

    return grp.reset_index().rename(columns={'index': 'date'})

print("2. Đang lấp đầy các ngày bị thiếu (có thể mất 1-2 phút)...")
# Áp dụng cho từng sản phẩm
df_final_perfect = df_fixed.groupby('product_id', group_keys=False).apply(fill_missing_dates)

print(f"Dữ liệu hiện có: {len(df_final_perfect):,}")

print("3. Đang tính lại Lag & Rolling trên chuỗi thời gian liên tục...")
group_prod = df_final_perfect.groupby('product_id')

df_final_perfect['day_of_week'] = df_final_perfect['date'].dt.dayofweek
df_final_perfect['is_weekend'] = df_final_perfect['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
df_final_perfect['view_lag_1'] = group_prod['luot_view'].shift(1)
df_final_perfect['view_lag_7'] = group_prod['luot_view'].shift(7)
df_final_perfect['mua_lag_1'] = group_prod['luot_mua'].shift(1)

df_final_perfect['view_rolling_7'] = group_prod['luot_view'].transform(lambda x: x.rolling(window=7, min_periods=7).mean())

df_model_final = df_final_perfect.dropna().reset_index(drop=True)

print(f"Danh sách cột chuẩn bị lưu: {df_model_final.columns.tolist()}")
display(df_model_final.head(5))

df_model_final.to_csv('/content/drive/MyDrive/Project2_Ecommerce/ecommerce_clean_data.csv', index=False)
print("ĐÃ LƯU FILE LÊN DRIVE!")
```
